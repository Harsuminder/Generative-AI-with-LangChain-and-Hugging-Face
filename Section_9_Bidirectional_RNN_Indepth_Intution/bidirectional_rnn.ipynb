## ðŸ”„ Bidirectional RNN (BiRNN)

A **Bidirectional RNN** is an architecture where the model processes the input sequence **in both forward and backward directions**. This allows the network to have information from both **past** and **future** contexts at every time step.

---

### ðŸ¤” Why Do We Need BiRNN?

In standard (unidirectional) RNNs:
- The model only looks at **past inputs** (left to right).
- But sometimes, the **future context** is also important for better understanding.

BiRNNs solve this by:
- Using **two RNN layers**:
  - One processes the sequence from start to end.
  - The other processes it from end to start.
- Their outputs are **concatenated** or **merged**, providing richer context.

---

### ðŸ“Œ Example: Fill-in-the-Blank (Middle Word Prediction)

> "**The cat ___ on the mat.**"

To correctly predict the blank word, we need:
- **Left context**: "The cat"
- **Right context**: "on the mat"

A standard RNN that only sees "The cat" might guess "**meows**" or "**runs**".

A **BiRNN**, seeing both sides, can better predict that the correct word is "**sat**", because it understands the relationship between **"cat"** and **"mat"** â€” which suggests a sitting or resting action.

With **BiRNN**, the model sees:
- What came **before** and
- What comes **after** each word.

This improves accuracy in tasks like:
- Machine translation
- NER
- Speech recognition
- Part-of-speech tagging
