## 🔁 Simple RNN

A **Simple RNN** is a neural network where outputs from previous time steps are fed back into the network as inputs. It learns patterns in sequential data, such as time series or text.

- Uses a **hidden state (`hₜ`)** to retain short-term memory.
- Struggles with **long-term dependencies**, which LSTM/GRU solve.

---

## 🔄 RNN Input-Output Architectures

### 1️⃣ **One-to-One**

- **Structure**: Single input → Single output
- **Example**: Image classification (e.g., classifying a photo as cat or dog)
- ✅ Standard neural network (non-RNN)

---

### 2️⃣ **One-to-Many**

- **Structure**: Single input → Sequence of outputs
- **Example**: Music generation from a mood or genre label
- 🎵 Input: "Jazz" → Output: Series of musical notes

---

### 3️⃣ **Many-to-One**

- **Structure**: Sequence of inputs → Single output
- **Example**: Sentiment analysis
- 📝 Input: "The movie was absolutely fantastic!" → Output: Positive

---

### 4️⃣ **Many-to-Many**

- **Structure**: Sequence of inputs → Sequence of outputs
- **Example**: Machine translation
- 🌍 Input: "Bonjour le monde" → Output: "Hello world"

---

🧠 These variations allow RNNs to handle different types of sequential learning problems across NLP, speech, music, and time-series domains.


## 🔄 Bidirectional RNN (BiRNN)

A **Bidirectional RNN** is an architecture where the model processes the input sequence **in both forward and backward directions**. This allows the network to have information from both **past** and **future** contexts at every time step.

---

### 🤔 Why Do We Need BiRNN?

In standard (unidirectional) RNNs:
- The model only looks at **past inputs** (left to right).
- But sometimes, the **future context** is also important for better understanding.

BiRNNs solve this by:
- Using **two RNN layers**:
  - One processes the sequence from start to end.
  - The other processes it from end to start.
- Their outputs are **concatenated** or **merged**, providing richer context.

---

### 📌 Example: Fill-in-the-Blank (Middle Word Prediction)

> "**The cat ___ on the mat.**"

To correctly predict the blank word, we need:
- **Left context**: "The cat"
- **Right context**: "on the mat"

A standard RNN that only sees "The cat" might guess "**meows**" or "**runs**".

A **BiRNN**, seeing both sides, can better predict that the correct word is "**sat**", because it understands the relationship between **"cat"** and **"mat"** — which suggests a sitting or resting action.

With **BiRNN**, the model sees:
- What came **before** and
- What comes **after** each word.

This improves accuracy in tasks like:
- Machine translation
- NER
- Speech recognition
- Part-of-speech tagging

---

### ✅ Summary

- **Standard RNN**: Only past context → 🔙
- **BiRNN**: Past + future context → 🔙🔜
- Improves performance on **sequence labeling and understanding tasks** by giving full visibility of the sequence at each time step.
