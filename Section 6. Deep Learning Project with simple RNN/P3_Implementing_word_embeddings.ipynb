```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = [
    "The movie was fantastic and full of suspense",
    "I did not enjoy the film at all",
    "What a wonderful performance by the lead actor",
    "The plot was weak and the characters were boring",
    "Absolutely loved the cinematography and music",
    "The story was predictable and unoriginal",
    "Great acting but the pacing was too slow",
    "A masterpiece with emotional depth and strong storytelling",
    "Terrible movie â€” waste of time",
    "The film had stunning visuals but lacked substance"
]
```

```python
# Define Vocabulary size

vocab_size=1000
```

```python
# One Hot Representation

one_hot_repr= [one_hot(words, vocab_size)for words in sent]
one_hot_repr
```

```python
# Word EMbedding Representation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_length=10
embedded_docs= pad_sequences(one_hot_repr, padding='pre', maxlen= sent_length)
embedded_docs
```

```python
# Feature Representation

dim=15
```

```python
model= Sequential()
model.add(Embedding(vocab_size, dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
# each word is represented by 15 dimensions
```
